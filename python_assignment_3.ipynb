```python
#8) Write a Python program to check the validity of password 
#input by users.
#At least 1 letter between [a-z] and 1 letter between [A-Z].
#At least 1 number between [0-9]. 
#At least 1 character from [$#@].
#Minimum length 6 characters
#Maximum length 16 characters.
password=input("enter a password")
l=len(password)
d_count=0
l_count=0
s_count=0
u_count=0
a_count=0
for i in password:
    if i.isalpha():
        a_count+=1
    if i.isdigit():
        d_count+=1
    if i.isupper():
         u_count+=1
    if i.islower():
         l_count+=1
    if i=="@" or i=="#" or i=="$":
         s_count+=1
if a_count>=1 and d_count>=1 and u_count>=1 and l_count>=1 and s_count>=1:
    print("valid password")
else:
    print("not valid password")
```

```python
#9) Write a program in Python to reverse a word by using 
#while loop
w=input("enter a word: ")
l=len(w)
s=1
k=""
while s<=l:
    k=k+w[l-s]
    s=s+1
print(k)
```

```python
#11) Write a program which takes integer inputs from user until 
#he/she presses “q” (Ask to press q to quit after every integer 
#input). Print average and product of all numbers
numbers=[]
newnum=[]
product=1
while True:
    num=input("give a number : ")
    if num=="q":
        break
    print("if you want to exit press - q")
    numbers.append(num)
for j in numbers:
    a=int(j)
    newnum.append(a)
for i in newnum:
    product=product*i
total=sum(newnum)
average=total/len(newnum)
print("\nProduct = {}".format(product))
print("Average = {}".format(average))
```
