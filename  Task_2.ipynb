```python
#5) Write a python program to convert all even indexed alphabets to upper and odd indexed char
def f1(s):
    le=len(s)
    l=""
    for j in range(0,le):
        if j%2==0:
            i=s[j].upper()
            l+=i
        else:
            h=s[j].lower()
            l+=h
    return l  
```

```python
#Write a python program to take three names as input from a user in the single input () 
#function call Example: Input: Enter three names: - “person1 person2 person3” 
#Output: Name1: - “person1” Name2: - “person2” Name3: - “person3”
def name(a):
    a=input("enter three names ")
    s=a.split()
    le=len(s)
    for j in range(le):
        if j==0:
            print("Name1: -",s[j])
        elif j==1:
            print("Name2: -",s[j])
        elif j==2:
            print("Name3: -",s[j])     
```

```python
name("person1 person2 person3")
```

```python
#10) Write a Python program to get a string from a given
#string where all occurrences of its first char have been changed to '@', except the first char itself.

s=input("enter your string : ")
s1=s[0]
s2=s[0]
l=len(s)
for i in range(1,l):
    if s[i]==s2:
        s1=s1 + "@"
    else:
        s1=s1+s[i]       
print(s1)   
```

```python
#19) Write a python program that takes a string, breaks it up and returns it with vowels first, consonants second. 
#For any character that's not a vowel 
#(like special characters or spaces), treat them like consonants

p=input("enter a string: ")
i=""
z=""
for s in p:
    if s=="a" or s=="e" or s=="i" or s=="o" or s=="u":
        i=i+s
    else:
        z+=s
print(i+z)
```

```python
#20) Create a dynamic calculator which asks for numbers and operator and return the answers

n1=int(input("enter any number :"))
o=input("enter any of these *,/,//,%,+,-,** : ")
n2=int(input("enter another number :"))
if o=="*":
    print(n1*n2)
elif o=="+":
    print(n1+n2)
if o=="//":
    print(n1//n2)
elif o=="**":
    print(n1**n2)
elif o=="/":
    print(n1/n2)
elif o=="%":
    print(n1%n2)
elif o=="-":
    print(n1-n2)
```
