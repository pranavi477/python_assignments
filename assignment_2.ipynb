```python
#write a python program to convert only odd index chars into lower case
#ex: PYTHON
#PyThOn
i=input()
e=""
for z in range(len(i)):
    if z%2!=0:
        s=i[z]
        e=e+s.lower()
    else:
        e=e+i[z]
print(e)        
        
```
